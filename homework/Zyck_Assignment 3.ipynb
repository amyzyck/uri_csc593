### Task

Copy your `Citation` class here from your class notes, then:

1. Write a function that takes the XML above and outputs a list of `Citation` objects containing the data from the XML. 
2. Run the function, loop over the `Citation`s, and prints the output of each one's `complete()` method.

```python
# Task 1: Write a function that takes the XML above and outputs a list of Citation objects containing the data from the XML.

import xml.etree.ElementTree as ET
xml_cite = ET.fromstring(xml)

def xml_citation(data):
    citations = []
    
    for cite in data:
        author = cite.find('author').text
        title = cite.find('title').text
        
        journal = cite.find('journal')
        if journal != None:
            journal = journal.text
            
        booktitle = cite.find('booktitle')
        if booktitle != None:
            booktitle = booktitle.text
            
        editor = cite.find('editor')
        if editor != None:
            editor = editor.text
        
        publisher = cite.find('publisher')
        if publisher != None:
            publisher = publisher.text
       
        volume = cite.find('volume')
        if volume != None:
            volume = volume.text
            
        pages = cite.find('pages')
        if pages != None:
            pages = pages.text
        
        location = cite.find('location')
        if location != None:
            location = location.text
        
        date = cite.find('date')
        if date != None:
            date = date.text

        citation = {
            "author" : author,
            "title" : title,
            "journal" : journal,
            "booktitle" : booktitle,
            "editor" : editor,
            "publisher" : publisher,
            "volume" : volume,
            "pages" : pages,
            "location" : location,
            "date" : date
        }
        
        citations.append(citation)
        
    return citations

xml_citation(xml_cite)
 
```

```python
# Task 2: Run the function, loop over the Citations, and prints the output of each one's complete() method.

input = xml_citation(xml_cite)

class Citation:
    def __init__(self, details):
        self.author = details['author']
        self.title = details['title']
        
        if 'journal' in details:
            self.journal = details['journal']
        else:
            self.journal = None
            
        if 'booktitle' in details:
            self.booktitle = details['booktitle']  
        else:
            self.booktitle = None
        
        if 'editor' in details:
            self.editor = details['editor']
        else:
            self.editor = None
        
        if 'publisher' in details:
            self.publisher = details['publisher']
        else:
            self.publisher = None
        
        if 'volume' in details:
            self.volume = details['volume']
        else:
            self.volume = None 
        
        if 'pages' in details:
            self.pages = details['pages']
        else:
            self.pages = None
        
        if 'location' in details:
            self.location = details['location']
        else:
            self.location = None
        
        if 'date' in details:
            self.date = details['date']
        else:
            self.date = None
        
    def names(self):
        return self.author.split(',')
    
    def complete(self):
        output = self.author + " " + self.title 
        if self.journal:
            output = output + self.journal
        if self.booktitle:
            output = output + self.booktitle
        if self.editor:
            output = output + self.editor
        if self.publisher:
            output = output + self.publisher
        if self.volume:
            output = output + " " + self.volume
        if self.pages:
            output = output + self.pages
        if self.location:
            output = output + self.location
        if self.date:
            output = output + " " + self.date
        print(output)
        
for n in range(len(input)):
    c = Citation(input[n])
    print(n+1, end = " ")
    c.complete()
    
```
