```python
#load a csv.
MSC = pd.read_csv('../dataset/MusselSpat_cleaned.csv')
MSC.head()
```

```python
bins = [136, 228, 274, 365]
MSC['Summer_Summary'] = pd.cut(MSC.Julian_Day, bins, 
                         labels=['Early Summer', 'Mid-Summer', 
                                 'Late Summer'])
MSC
```

```python
bins1 = [0, 0.75 , 1.5, 3, 10]
MSC['Size_Class'] = pd.cut(MSC.Size, bins1, 
                         labels=['Small', 'Medium', 
                                 'Large','Late Juvenile'])
MSC
```

```python
bins2 = [-1, 0.5, 1]
MSC['Eyespot_Class'] = pd.cut(MSC.Eyespot, bins2, 
                         labels=['Recently settled spat', 'Settled over 5 days ago'])
MSC
```
